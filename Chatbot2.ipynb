```python
%pip install transformers torch gradio

import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# FREE MODEL (Not Microsoft)
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float32,
    device_map="cpu"   # Use 'cuda' if you have GPU
)

def explain_topic(topic):
    prompt = f"<|system|> You are a helpful AI tutor. Explain the topic clearly.\n<|user|> Explain: {topic}\n<|assistant|>"

    inputs = tokenizer(prompt, return_tensors="pt")

    output = model.generate(
        **inputs,
        max_new_tokens=200,
        temperature=0.5,
        do_sample=True,
    )

    text = tokenizer.decode(output[0], skip_special_tokens=True)

    # Clean response
    if "<|assistant|>" in text:
        text = text.split("<|assistant|>")[-1].strip()

    return text

interface = gr.Interface(
    fn=explain_topic,
    inputs="text",
    outputs="text",
    title="AI Explainer (TinyLlama)",
    description="Type a topic and the AI will explain it clearly."
)

interface.launch()
```
